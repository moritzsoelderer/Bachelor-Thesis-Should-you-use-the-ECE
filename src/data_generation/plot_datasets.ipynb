```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D

from src.data_generation.datasets import exclamation_mark_dataset, gummy_worm_dataset
```

```python
num_samples_per_dist = 10000
```

```python
X, y = gummy_worm_dataset().generate_data(num_samples_per_dist)
```

```python
colormap = np.array(['red', 'blue'])

plt.figure(figsize=(10, 10), dpi=200)
plt.title('Gummy Worm Dataset', fontsize=24)

plt.scatter(X[:, 0], X[:, 1], c=colormap[y], s=0.8)
plt.grid(True, linestyle='--', alpha=0.6)

legend_elements = [
Line2D([0], [0], marker='o', color='white', markerfacecolor=colormap[label], label=str(label))
for label in np.unique(y)
]
legend = plt.legend(handles=legend_elements, title="Classes", loc='upper left', fontsize=14)
legend.get_title().set_fontsize(14)

plt.xlabel("feature 0", fontsize=16)
plt.ylabel("feature 1", fontsize=16)

plt.savefig('./plots/datasets/gummy_worm_dataset.png')
```

```python
X_ex, y_ex = exclamation_mark_dataset().generate_data(num_samples_per_dist)
```

```python
colormap = np.array(['red', 'blue'])

plt.figure(figsize=(10, 10), dpi=200)
plt.title('Exclamation Mark Dataset', fontsize=24)

plt.scatter(X_ex[:, 0], X_ex[:, 1], c=colormap[y_ex], s=0.8)
plt.grid(True, linestyle='--', alpha=0.6)

legend_elements = [
Line2D([0], [0], marker='o', color='white', markerfacecolor=colormap[label], label=str(label))
for label in np.unique(y_ex)
]
legend = plt.legend(handles=legend_elements, title="Classes", loc='upper left', fontsize=14)
legend.get_title().set_fontsize(14)

plt.xlabel("feature 0", fontsize=16)
plt.ylabel("feature 1", fontsize=16)

plt.savefig('./plots/datasets/exclamation_mark_dataset.png')
```
